<h3> Packages and modules </h3>

- what is package X module
- module = org. unit of python code
    - it is NOT a single file
    - it HAS NAMESPACE - name_of_module.thing_in_the_module (e.g. csv.reader)
- package is also a module
    - and can contain other modules or packages
    - technically it is a module with ___path___ attr
    - e.g. PyGal is a package
- example: it is all about organization
    - import animals ... imports whole package
    - import animals.lizards ... imports a submodule of lizards
    - import animals.lizards.chameleon ... imports an object from submodule
- all packages are modules but not all modules are packages
- so packages/modules are not files per se, but can be stored like that
- ... thanks to it I can make my own modules easy just by saving .py files
    - e.g. file filled with function for crypto calculations
    - saved as crypto_calculations.py
    - ... now I can import them by: import crypto_calculations as crptcalc
    - now I can use them!
    - *the module Im importing just has to be in the same directory

<h3> Line plots with Pygal </h3>

- conventional plots
- x and y scatter plots

```python
import pygal

def draw_line(title, xvals, yvals):
    """
    Draw line plot with given x and y values.
    """
    lineplot = pygal.Line(height = 400)   #not necessary, but easier to show
    lineplot.tile = title
    lineplot.x_labels = xvals
    lineplot.add("Data", yvals)
    display(lineplot)

def draw_xy(title, xvals, yvals):
    """
    Draw xy plot with given x and y values.
    """
    coords = [(xval, yval) for xval, yval in zip(xvals, yvals)]     #zip takes two sequences into for looping
    xyplot = pygal.XY(height = 400)
    xyplot.title = title
    xyplot.add("Data", coords)
    display(xyplot)

xvals = [0,1,3,5,6,7,9,11,12,15]
yvals = [4,3,1,2,2,4,5,2,1,4]
draw_line("Line plot", xvals, yvals)
draw_xy("XY plot", xvals, yvals)
#the diff between line plot and XY plot is the assumption of same differences between values
#in line plot the xvals are rather categories (even though they are numbers)
#in XY plot it can tell there are actually numbers (e.g ints)
```
